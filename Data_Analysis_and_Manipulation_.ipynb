# **Numpy**

```python
import numpy as np
```

```python
## A 1-D array called zeros having 10 elements and all the elements are set to zero.

zeroarray= np.zeros(10)
zeroarray
```

```python
#A 1-D array called vowels having the elements ‘a’, ‘e’, ‘i’, ‘o’ and ‘u’.

vowelarray=np.array(['a', 'e', 'i', 'o','u'])
vowelarray
```

```python
# A 2-D array called ones having 2 rows and 5 columns and all the elements are set to 1 and dtype as int.

onestwodarray=np.ones(10,dtype=int).reshape(2,5)
onestwodarray
```

```python
# Find the dimensions, shape, size, data type of the items and itemsize of arrays zeros, vowels, ones, myarray1 and myarray2.



#Attribute of zeros array
print("Attribut of zeros Array:")
print("Dimension: ",zeroarray.ndim)
print("Shape: ",zeroarray.shape)
print("Size ",zeroarray.size)
print("Data Type: ",zeroarray.dtype)

#Attribute of Vowels array
print("\nAttribut of Vowels Array:")
print("Dimension: ",vowelarray.ndim)
print("Shape: ",vowelarray.shape)
print("Size ",vowelarray.size)
print("Data Type: ",vowelarray.dtype)

#Attribute of Ones array
print("\nAttribut of Ones Array:")
print("Dimension: ",onestwodarray.ndim)
print("Shape: ",onestwodarray.shape)
print("Size ",onestwodarray.size)
print("Data Type: ",onestwodarray.dtype)

#Attribute of myarray1 array
print("\nAttribut of myarray1 Array:")
print("Dimension: ",myarray1.ndim)
print("Shape: ",myarray1.shape)
print("Size ",myarray1.size)
print("Data Type: ",myarray1.dtype)

#Attribute of myarray2 array
print("\nAttribut of myarray2 Array:")
print("Dimension: ",myarray2.ndim)
print("Shape: ",myarray2.shape)
print("Size ",myarray2.size)
print("Data Type: ",myarray2.dtype)
```

# **Pandas**

```python
import pandas as pd
```

```python
# Create a pandas Series S1 of elements [2, 5, 6, 8, 5, -10, 2, -2].

s1=pd.Series([2, 5, 6, 8, 5, -10, 2, -2])
s1
```

```python
# Recreate a pandas Series S1, include the index option and assign an array of strings containing the labels.

s1= pd.Series([2, 5, 6, 8, 5, -10, 2, -2],index=['a','b','c','d','e','f','g','h'])
s1
```

```python
# Display two arrays individually that make up the Series data structure in Pandas using index and values attributes.

index=s1.index
print('Index:',index)

values=s1.values
print("Values: ",values)
```

```python
# Select individual elements of a Series S1 by specifying the Key and Label corresponding to the position of the index.

print(s1[2])
print(s1['c'])
```

```python
# Select multiple elements of a Series S1 by specifying the Key and Label corresponding to the position of the index using slicing.

print(s1[0:5])

print(s1['a':'e'])
```

```python
#  Create a Series S2 and S3 from NumPy Arrays and any other Series.

np1=np.array([2,4,1,5,6,8,3])
s2=pd.Series(np1)
print("S2: ",s2)
np2=np.array([5,-1,0-2,-3,5,4,6])
s3=pd.Series(np2)
print('S3: ',s3)
```

```python
# Select all elements of Series S1 which are less than 8, Series S2 which are greater than 5 and Series S3 which are less than 0.

print("S1 <8: ")
print(s1[s1<8])

print("S2 >5: ")
print(s2[s2>5])

print("S3 <0 ")
print(s3[s3<0])
```

```python
# Perform operations using operators such as (+, -, *, and /) on Series S1, S2 and S3,and any three mathematical functions on each Series.


print(s1+2)
print(s1/2)
print(s1*2)
print(s1-2)

print(s2+5)
print(s2-5)
print(s2*5)
print(s2/5)

print(s3+7)
print(s3-7)
print(s3*7)
print(s3/7)

#########################Mathemetical function

print(s1.pow(2))
print(s2.pow(3))
print(s3.pow(4))

print("S1 sum:" ,s1.sum())
print("S2 sum:" ,s2.sum())
print("S3 sum:" ,s3.sum())

print("ABS s1: ")
print(s1.abs())

print("ABS s2: ")
print(s2.abs())

print("ABS s3: ")
print(s3.abs())
```

```python
# Select all the values contained in the Series S1, excluding duplicate.

s1.unique()
```

```python
# Select all the values contained in the Series S1, excluding duplicates and count the occurrences of each unique value.

s1.value_counts()

```

```python
# Demonstrate the use of isin() membership function on each of the Series S1, S2 and S3.


#check which index contains values from [2,4,6,8,10]
print(s1.isin([2,4,6,8,10]))

#check which index contains values from [1,2,3]
print(s2.isin([1,2,3]))

#check which index contains 0

print(s3.isin([0]))
```

```python
# Create a Series S4 of elements [2, 5, 6, np.NaN, 5, np.NaN, 2, -2] and demonstrate the use of isnull() and notnull() functions on Series S4.


s4=pd.Series([2, 5, 6, np.NaN, 5, np.NaN, 2, -2] )

#use of isnull()
print(s4.isnull())

#use of notnull()
print(s4.notnull())
```
